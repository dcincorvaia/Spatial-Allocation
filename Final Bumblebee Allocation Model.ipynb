```python
# this cell will be used to figure out probability of a naive bee finding a resource
def MakeLandscape(rows,columns,resource_space, resource_number):
    myGrid = []
    resources = list(range(resource_space))
    for n in range(rows):
        myRow = []
        for n in range(columns):
            myRow.append(0)
        myGrid.append(myRow)
    for n in range(resource_number):
        for n in range(len(resources)):
            random_row = numpy.random.randint(rows)
            random_column = numpy.random.randint(columns)
            while myGrid[random_row][random_column] != 0:
                random_row = numpy.random.randint(rows)
                random_column = numpy.random.randint(columns)
            if myGrid[random_row][random_column] == 0:
                myGrid[random_row][random_column] = resources[n] + 1
    return(myGrid)

def MakeBees(N):
    Bees = []
    for n in range(N):
        Bees.append(0)
    return(Bees)

def AverageLists(list_of_lists,divisor):
    final_list = list()
    position = 0
    while position < len(list_of_lists[0]):
        values = list()
        list_number = 0
        while list_number < len(list_of_lists):
            values.append(list_of_lists[list_number][position])
            list_number = list_number + 1
        position = position + 1
        final_list.append((sum(values)/divisor))
    return final_list

info_prob = 0
movements = [1,0,-1]
x = 0
listOfZeroes = list()
while x < 2000:
    Rows = 100
    Columns = 100
    Resource_Range = 5
    Resource_Number = 2
    Bee = 100
    random_search = 100
    search_prob = 0.01
    Trial_Length = 1
    Trial_Number = 1
    Resource_total = list()
    for n in list(range(Trial_Number)):
        Bees = MakeBees(Bee)
        Best_Resource = 0
        Resources = 0
        Landscape = MakeLandscape(Rows,Columns, Resource_Range,Resource_Number)
        for n in list(range(Trial_Length)):
            for N in range(len(Bees)):
                ## this if statement covers search if bee N's resource is worse than the best resource
                if Best_Resource > Bees[N] and Bees[N] != 0 and numpy.random.random() < info_prob:
                    time = 0
                    row = numpy.random.randint(Rows)
                    column = numpy.random.randint(Columns)
                    position = Landscape[row][column] 
                    while time < random_search:
                        if position > Bees[N]:
                            Bees[N] = position
                            time = random_search
                            Resources = Resources + Bees[N]
                            if Bees[N] > Best_Resource:
                                Best_Resource = Bees[N]
                        else:
                            try: 
                                new_row = row + numpy.random.choice(movements)
                                new_column = column + numpy.random.choice(movements)
                                new_position = Landscape[new_row][new_column]
                            except:
                                new_position = "NA"
                            while new_position == "NA":
                                try: 
                                    new_row = row + numpy.random.choice(movements)
                                    new_column = column + numpy.random.choice(movements)
                                    new_position = Landscape[new_row][new_column]
                                except:
                                    new_position = "NA"
                            time = time + 1
                            row = new_row
                            column = new_column
                            position = new_position


                else: 
                    ## this covers if the bee hasn't found any resource yet (automatically searches)
                    if Bees[N] == 0:
                            time = 0
                            row = numpy.random.randint(Rows)
                            column = numpy.random.randint(Columns)
                            position = Landscape[row][column] 
                            while time < random_search:
                                if position > Bees[N]:
                                    Bees[N] = position
                                    time = random_search
                                    Resources = Resources + Bees[N]
                                    if Bees[N] > Best_Resource:
                                        Best_Resource = Bees[N]
                                else:
                                    try: 
                                        new_row = row + numpy.random.choice(movements)
                                        new_column = column + numpy.random.choice(movements)
                                        new_position = Landscape[new_row][new_column]
                                    except:
                                        new_position = "NA"
                                    while new_position == "NA":
                                        try: 
                                            new_row = row + numpy.random.choice(movements)
                                            new_column = column + numpy.random.choice(movements)
                                            new_position = Landscape[new_row][new_column]
                                        except:
                                            new_position = "NA"
                                    time = time + 1
                                    row = new_row
                                    column = new_column
                                    position = new_position

                    if Bees[N] > 0:
                        ## this if statement covers random search
                        if numpy.random.random() < search_prob:
                                time = 0
                                row = numpy.random.randint(Rows)
                                column = numpy.random.randint(Columns)
                                position = Landscape[row][column]
                                while time < random_search:
                                    if position > Bees[N]:
                                        Bees[N] = position
                                        time = random_search
                                        Resources = Resources + Bees[N]
                                        if Bees[N] > Best_Resource:
                                            Best_Resource = Bees[N]
                                    else:
                                        try: 
                                            new_row = row + numpy.random.choice(movements)
                                            new_column = column + numpy.random.choice(movements)
                                            new_position = Landscape[new_row][new_column]
                                        except:
                                            new_position = "NA"
                                        while new_position == "NA":
                                            try: 
                                                new_row = row + numpy.random.choice(movements)
                                                new_column = column + numpy.random.choice(movements)
                                                new_position = Landscape[new_row][new_column]
                                            except:
                                                new_position = "NA"
                                        time = time + 1
                                        row = new_row
                                        column = new_column
                                        position = new_position
                        else:
                            Resources = Resources + Bees[N]                
        listOfZeroes.append(Bees.count(0))
        x = x+1
#print(listOfZeroes)
print(average(listOfZeroes))
```
